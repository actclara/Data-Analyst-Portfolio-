```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt 
```

# Load 
upload 12 months of cycling data from 07-2020 to 06-2021

```python
#upload 12 months of cycling data 
df_1 = pd.read_csv('Cyclistic 12 months data/202007-divvy-tripdata.csv')
df_2 = pd.read_csv('Cyclistic 12 months data/202008-divvy-tripdata.csv')
df_3 = pd.read_csv('Cyclistic 12 months data/202009-divvy-tripdata.csv')
df_4 = pd.read_csv('Cyclistic 12 months data/202010-divvy-tripdata.csv')
df_5 = pd.read_csv('Cyclistic 12 months data/202011-divvy-tripdata.csv')
df_6 = pd.read_csv('Cyclistic 12 months data/202012-divvy-tripdata.csv')
df_7 = pd.read_csv('Cyclistic 12 months data/202101-divvy-tripdata.csv')
df_8 = pd.read_csv('Cyclistic 12 months data/202102-divvy-tripdata.csv')
df_9 = pd.read_csv('Cyclistic 12 months data/202103-divvy-tripdata.csv')
df_10 = pd.read_csv('Cyclistic 12 months data/202104-divvy-tripdata.csv')
df_11 = pd.read_csv('Cyclistic 12 months data/202105-divvy-tripdata.csv')
df_12 = pd.read_csv('Cyclistic 12 months data/202106-divvy-tripdata.csv')
```

# Cleaning Phase 

- dropping duplicates in all dataframes 
- convert 'started_at' and 'ended_at' to appropriate datetime format
- calculate ride length and delete ride length which are less than 00:00:00 because it's impossible to have
a negative ride length 
- Change day of week to numeric variables. Monday = 0 and Sunday = 6

```python
#merging all 12 dataframes into one and removing duplicates 
df = [df_1,df_2,df_3,df_4,df_5,df_6,df_7,df_8,df_9,df_10,df_11,df_12]
df_final = pd.concat(df).drop_duplicates()
df_final.head()
```

```python
df_final['month'] = pd.DatetimeIndex(df_final['started_at']).month
df_final['year'] = pd.DatetimeIndex(df_final['started_at']).year
```

```python
#obtain the list of dataframe column ['started_at']
#Obtain the list of dataframe column ['ended_at']
#convert to datetime format 
#when a csv file is imported and a dataframe is made, datatime objects in the file
#are read as strings and difficult to perform operations such as time difference 

df_final['started_at']=pd.to_datetime(pd.Series(df_final['started_at']))
df_final['ended_at']=pd.to_datetime(pd.Series(df_final['ended_at']))

#subtract started_at from ended_at 
#find the ride_length on each dataframe 
df_final['ride_length']=df_final['ended_at']-df_final['started_at']

#Monday = 0 
#Sunday = 6 
df_final['day_of_week']=df_final['started_at'].dt.dayofweek 


#removing ride_lengths that have negative values 
df_final = df_final[df_final['ride_length'] >= '0 days']

df_final.head()
```

# Analyze Phase 

###  Obtain the summary ride lengths from annual membership riders and casual riders 
- Which day(s) and month(s) do casual riders and annual membership riders prefer to cycle? 
- What's the average ride length of casual riders and annual membership riders? 
- What's the peak time of cycling for casual riders and annual membership riders? 
- Total number of ride lengths in each day of casual riders and annual membership riders?
- Favourite spot to ride? 

Once the questions above are answered, results are exported as a csv file to upload in Google Sheet to create data visualizations of the analyse phase. 

```python
#members 
members_df = df_final[df_final['member_casual']=='member']
#casual members 
casual_df = df_final[df_final['member_casual']=='casual']
#print(casual['day_of_week'].value_counts().to_frame().idxmax())
#print(casual['day_of_week'].value_counts().idxmax())
name_of_month=['01-2021','02-2021','03-2021','04-2021','05-2021','06-2021','07-2020','08-2020','09-2020','10-2020','11-2020','12-2020']
#create dataframe 
casual_day = pd.DataFrame(name_of_month,columns=['Month_Year'])
value = [] 
#Monday = 0 
#Sunday = 6 
day_of_week = ['Monday','Tuesday','Wednesday','Thursday','Friday','Saturday','Sunday']

for month in range(1,13): 
    #if month=1, then we are analysing for January 2021
    casual = casual_df[casual_df['month']==month]  
    #if month=1, which day of week in Jan has most Cyclistic casual users 
    result = casual['day_of_week'].value_counts().idxmax() 
    value.append(day_of_week[result])
    #print ('Favourite Day of the week casual users use Cyclistic the most in', name_of_month[month-1],'is' ,day_of_week[result])

#inserting the result into the casual_day dataframe
casual_day['Day']=value 
casual_day['Membership']='Casual'

```

```python
#for Cyclistic members 
member_day = pd.DataFrame(name_of_month,columns=['Month_Year'])
value = []
for month in range(1,13): 
    members = members_df[members_df['month']==month]
    result = members['day_of_week'].value_counts().idxmax()
    value.append(day_of_week[result])
    #print ('Favourite Day of the week members use Cyclistic the most in', name_of_month[month-1],'is' ,day_of_week[result])

member_day["Day"]=value
member_day["Membership"]="Member"
```

```python
#Mean ride_lengths per months from casual members 


#casual_rider_length = casual_df[casual_df['month']==7]
#casual_rider_length['ride_length'].describe()
value = []
#Average of ride lengths per month 
for month in range(1,13): 
    casual_month = casual_df[casual_df['month']==month]
    casual_rider_mean_length = casual_month['ride_length'].mean()
    value.append(casual_rider_mean_length)
    #print("Average Ride Length of Casual Member on",name_of_month[month-1],":",casual_rider_mean_length)


casual_day["Average Ride Length"]=value 
```

```python

#Average of ride lengths per month of members
value = []
for month in range(1,13): 
    member_month = members_df[members_df['month']==month]
    member_rider_mean_length = member_month['ride_length'].mean()
    value.append(member_rider_mean_length)
    #print("Average Ride Length of Casual Member on",name_of_month[month-1],":",member_rider_mean_length)

member_day["Average Ride Length"]=value
```

```python
#Total ride length per month from casual members
value = []
for month in range(1,13): 
    casual_month = casual_df[casual_df['month']==month]
    casual_rider_total_length = casual_month['ride_length'].sum()
    value.append(casual_rider_total_length)
    #print("Total Ride Length of Casual Member on",name_of_month[month-1],":",member_rider_total_length)
    
casual_day["Total Ride Length"]=value
```

```python
#Total ride length per month from  member 
value = []
for month in range(1,13): 
    member_month = members_df[members_df['month']==month]
    member_rider_total_length = member_month['ride_length'].sum()
    value.append(member_rider_total_length)
    #print("Total Ride Length of Casual Member on",name_of_month[month-1],":",member_rider_total_length)
    
member_day["Total Ride Length"]=value
```

```python
#Most frequent start station name of members
value = [] 
members_df['start_station_name'].fillna(0)
members_df['start_station_name'].value_counts()

for month in range(1,13): 
    member_month = members_df[members_df['month']==month]
    member_start_station = member_month['start_station_name'].value_counts().idxmax()
    value.append(member_start_station)
    #print("Top start station of members",name_of_month[month-1],":",member_start_station)
    
member_day["Top Start Station"]=value 
```

```python
#Most frequent end station name of members

members_df['end_station_name'].fillna(0)
members_df['end_station_name'].value_counts()
value = []

for month in range(1,13): 
    member_month = members_df[members_df['month']==month]
    member_end_station = member_month['end_station_name'].value_counts().idxmax()
    value.append(member_end_station)
    #print("Top end station of members",name_of_month[month-1],":",member_end_station)
    
member_day["Top End Station"]=value
```

```python
#Most frequent start station name of casual members
casual_df['start_station_name'].fillna(0)
casual_df['start_station_name'].value_counts()
value = []
for month in range(1,13): 
    casual_month = casual_df[casual_df['month']==month]
    casual_start_station = casual_month['start_station_name'].value_counts().idxmax()
    value.append(casual_start_station)
    #print("Top start station on of casual members",name_of_month[month-1],":",casual_start_station)

casual_day["Top Start Station"]=value
```

```python
#Most frequent end station name of casual members

casual_df['end_station_name'].fillna(0)
casual_df['end_station_name'].value_counts()
value = []
for month in range(1,13): 
    casual_month = casual_df[casual_df['month']==month]
    casual_end_station = casual_month['end_station_name'].value_counts().idxmax()
    value.append(casual_end_station)
    #print("Top end station of casual members",name_of_month[month-1],":",casual_start_station)

casual_day["Top End Station"]=value 
```

```python
#casual_df['peak_hour']=casual_df['started_at'].dt.hour
casual_df_copy = casual_df.copy()
casual_df_copy['time_stamp_hour'] = casual_df_copy['started_at'].astype(str).str[10:13]
casual_df = casual_df_copy
#casual_df.head()
```

```python
#Finding peak hour for casual member 

value = []
total = []
for month in range(1,13): 
    casual_month = casual_df[casual_df['month']==month]
    casual_peak_hour = casual_month['time_stamp_hour'].value_counts().idxmax()
    total_number = max(casual_month['time_stamp_hour'].value_counts())
    value.append(casual_peak_hour)
    total.append(total_number)
    #print("Casual Member Peak hour on ",name_of_month[month-1],":",casual_peak_hour,":00:00","Number:",total_number)

casual_day["Peak Hour Time"]=value
casual_day["Number at Peak Hour"]=total
#casual_day.head()
```

```python
#combining both dataframes together 
frames = [casual_day,member_day]
final = pd.concat(frames,ignore_index=True,sort=False)
final[["Month","Year"]]=final.Month_Year.str.split("-",expand=True)
final = final.drop('Month_Year',axis=1)
first_column = final.pop('Year')
second_column = final.pop('Month')
final.insert(0,'Year',first_column)
final.insert(1,'Month',second_column)
final['Peak Hour Time'] = final['Peak Hour Time'].astype(str) + ":00"
final = final.sort_values(by=['Year','Month'])
```
