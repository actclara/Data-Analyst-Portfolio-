# Cleaning Phase 

- dropping duplicates in all dataframes 
- convert 'started_at' and 'ended_at' to appropriate datetime format
- calculate ride length and delete ride length which are less than 00:00:00 because it's impossible to have
a negative ride length 
- Change day of week to numeric variables. Monday = 0 and Sunday = 6

```python
#obtain the list of dataframe column ['started_at']
#Obtain the list of dataframe column ['ended_at']
#convert to datetime format 
#when a csv file is imported and a dataframe is made, datatime objects in the file
#are read as strings and difficult to perform operations such as time difference 

df_1['started_at']=pd.to_datetime(pd.Series(df_1['started_at']))
df_2['started_at']=pd.to_datetime(pd.Series(df_2['started_at']))
df_3['started_at']=pd.to_datetime(pd.Series(df_3['started_at']))
df_4['started_at']=pd.to_datetime(pd.Series(df_4['started_at']))
df_5['started_at']=pd.to_datetime(pd.Series(df_5['started_at']))
df_6['started_at']=pd.to_datetime(pd.Series(df_6['started_at']))
df_7['started_at']=pd.to_datetime(pd.Series(df_7['started_at']))
df_8['started_at']=pd.to_datetime(pd.Series(df_8['started_at']))
df_9['started_at']=pd.to_datetime(pd.Series(df_9['started_at']))
df_10['started_at']=pd.to_datetime(pd.Series(df_10['started_at']))
df_11['started_at']=pd.to_datetime(pd.Series(df_11['started_at']))
df_12['started_at']=pd.to_datetime(pd.Series(df_12['started_at']))

df_1['ended_at']=pd.to_datetime(pd.Series(df_1['ended_at']))
df_2['ended_at']=pd.to_datetime(pd.Series(df_2['ended_at']))
df_3['ended_at']=pd.to_datetime(pd.Series(df_3['ended_at']))
df_4['ended_at']=pd.to_datetime(pd.Series(df_4['ended_at']))
df_5['ended_at']=pd.to_datetime(pd.Series(df_5['ended_at']))
df_6['ended_at']=pd.to_datetime(pd.Series(df_6['ended_at']))
df_7['ended_at']=pd.to_datetime(pd.Series(df_7['ended_at']))
df_8['ended_at']=pd.to_datetime(pd.Series(df_8['ended_at']))
df_9['ended_at']=pd.to_datetime(pd.Series(df_9['ended_at']))
df_10['ended_at']=pd.to_datetime(pd.Series(df_10['ended_at']))
df_11['ended_at']=pd.to_datetime(pd.Series(df_11['ended_at']))
df_12['ended_at']=pd.to_datetime(pd.Series(df_12['ended_at']))
```

```python
#subtract started_at from ended_at 

#find the ride_length on each dataframe 
df_1['ride_length']=df_1['ended_at']-df_1['started_at']
df_2['ride_length']=df_2['ended_at']-df_2['started_at']
df_3['ride_length']=df_3['ended_at']-df_3['started_at']
df_4['ride_length']=df_4['ended_at']-df_4['started_at']
df_5['ride_length']=df_5['ended_at']-df_5['started_at']
df_6['ride_length']=df_6['ended_at']-df_6['started_at']
df_7['ride_length']=df_7['ended_at']-df_7['started_at']
df_8['ride_length']=df_8['ended_at']-df_8['started_at']
df_9['ride_length']= df_9['ended_at']-df_9['started_at']
df_10['ride_length']=df_10['ended_at']-df_10['started_at']
df_11['ride_length']=df_11['ended_at']-df_11['started_at']
df_12['ride_length']=df_12['ended_at']-df_12['started_at']
```
